7) Compound Interest: 
Write a Python program to calculate compound interest. Use the formula:
A = Px(1+r/2)**n-t
Where: 
A= total amount 
P = principal amount 
r = annual interest rate (decimal) 
n = number of times interest is compounded per year 
t = time in years 
Take P, r, n, and t as inputs from the user.

```python
p = int(input("Enter principal amount: "))
r = float(input("Enter annual interest rate (decimal): "))
n = int(input("Enter number of times interest is compounded per year: "))
t = int(input("Enter time in years: "))
a = p*(1+r/n)**(n-t)
print("The Compount of Interest is: ",a)
```

19) Work Done by a Force:
Write a Python program to calculate the work done by a force. Use the formula: 
where: 
W = work done 
f= force 
d = distance 
theta = angle between force and direction of movement (in degrees) 
Take f, d, and Î¸ as inputs from the user.

```python
import math

f = float(input("Enter force (Newtons): "))
d = float(input("Enter distance (meters): "))
theta = float(input("Enter angle (degrees): "))

theta_rad = math.radians(theta)

work = f * d * math.cos(theta_rad)

print("Work Done =", work, "Joules")
```
