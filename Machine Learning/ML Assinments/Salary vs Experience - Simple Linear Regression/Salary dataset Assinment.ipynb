### Q: Is assignment mein aap Salary Data dataset par Simple Linear Regression apply karein ge jahan experience ki base par salary predict ki jaye gi. Notebook mein data preprocessing, model training, evaluation aur prediction include honi chahiye.

```python
import pandas as pd
```

```python
data = pd.read_csv("Salary_Data.csv")
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
x = data[["YearsExperience"]]
y = data[["Salary"]]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# X = features, y = target labels
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
salary_predictor = LinearRegression()
```

```python
salary_predictor.fit(X_train,y_train)
```

```python
salary_predictor.coef_
```

```python
salary_predictor.intercept_
```

```python
import matplotlib.pyplot as plt
```

```python
data.plot(kind='scatter', x='YearsExperience', y='Salary', title='Salary Based on Experienced')
```

```python
y_pred_train = salary_predictor.predict(X_train)
```

```python
plt.scatter(x=X_train,y=y_train, color='yellow')
plt.scatter(x=X_train,y=y_pred_train, color='red')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_train, y_pred_train)
```

```python
r2_score(y_train, y_pred_train)
```

```python
y_pred_test = salary_predictor.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred_test)
```

```python
r2_score(y_test, y_pred_test)
```
