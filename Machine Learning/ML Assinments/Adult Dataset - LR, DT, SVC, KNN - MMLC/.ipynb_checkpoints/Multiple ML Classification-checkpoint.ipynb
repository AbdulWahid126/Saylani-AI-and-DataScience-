### 1. Work on Logistic Regression:


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import  StandardScaler
```

```python
data = pd.read_csv('adult.csv')
data.head()
```

```python
data.info()
```

```python
data.duplicated().sum()
```

```python
data.workclass.unique()
```

```python
data.income.unique()
```

```python
X = data.drop("income", axis=1)
y = data.income
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42)
```

```python
cat_cols = X_train.select_dtypes('object')
num_cols = X_train.select_dtypes('number')
```

```python
cat_cols
```

```python
num_cols
```

```python
model  = LogisticRegression()
```

```python
#  Numerical pipeline
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),   # handle missing numbers
    ("scaler", StandardScaler())
])

# Categorical pipeline
cat_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),  # handle missing categories
    ("encoder", OneHotEncoder(drop="first", handle_unknown="ignore",sparse_output=False))
])
```

```python
preprocessor = ColumnTransformer(
                            transformers=[
                            ("num", num_pipeline, num_cols.columns),
                            ("cat", cat_pipeline, cat_cols.columns)
                            ])
```

```python
model_pipeline = Pipeline(
    [
    ("preprocessor",  preprocessor),
    ("model", model)
    ]
)

model_pipeline.fit(X_train, y_train)
```
