# Simple Linear Regression

```python
import pandas as pd
```

```python
data = pd.read_csv("weight-height.csv")
```

```python
data.head()
```

```python
data.Gender.replace(['Male','Female'],[1,0], inplace=True)
```

```python
data.corr()
```

# Linear Regression
## Simple Linear Regression
    - 1 input feature
### y = mx + b
### weight = m(Height)+bias

```python
X = data[["Height"]]
y = data[["Weight"]]
```

```python
from sklearn.model_selection import train_test_split

# Splits data into 75% training and 25% testing (default)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

# Weight = 7.70 (Height) + (-349.78)

```python
y_pred_train = model.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from math import sqrt
sqrt(mean_squared_error(y_train, y_pred_train))
```

```python
r2_score(y_train, y_pred_train)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=X_train, y=y_train)
plt.scatter(x=X_train, y=y_pred_train)
```

```python
y_pred_test = model.predict(X_test)
```

```python
sqrt(mean_squared_error(y_test, y_pred_test))
```

```python
r2_score(y_test, y_pred_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=X_test, y=y_test)
plt.scatter(x=X_test, y=y_pred_test)
```

# Overfiting:
