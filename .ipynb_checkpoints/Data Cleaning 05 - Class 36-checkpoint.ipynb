# Replaceing Values

```python
import pandas as pd
```

```python
data = pd.read_excel('employee.xlsx')
```

```python
data
```

```python
import numpy as np
data.Department.replace('Account', 'Accounts')
data.Department.replace(['Account', 'Production',np.nan],['Account', 'Development', 50000])
```

# Descreatization and Binning:
 - Numbers:
    - Descreate Number
      - Countable [2,3,5,6]
    - Continous Number
      - Measurable [2,3,4,5, 6.5,5.3, 7.5]

```python
bins = [20,30,40,50]
data['Age_Category'] = pd.cut(data.Age, bins, labels=['Young', 'Adult', 'Senior'])
```

```python
data
```

```python
data.Age_Category.value_counts()
```

```python
import seaborn as sns
sns.barplot(data.Age_Category.value_counts())
```
